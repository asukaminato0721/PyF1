# Examples

## Basic Examples

```python
from sympy import *
t=symbols("t")
fourier_series(t/2,(t,-pi,pi)).truncate(3).rewrite(exp)
```

```python
from sympy.abc import *
from sympy import *
plot(fourier_series(t/2,(t,-pi,pi)).truncate(3).rewrite(exp),(t,-3*pi,3*pi))
```

```python
from sympy.abc import *
from sympy import *
fourier_series(x**2*y,(x,-pi,pi),(y,-pi,pi)).truncate(1).subs(x,1).subs(y,1).evalf()
# I don't know multi-dimension of sympy
```

```python
from sympy.abc import *
from sympy import *
from sympy.plotting import *
plot3d(fourier_series(x**2*y,(x,-pi,pi),(y,-pi,pi)).truncate(2),(x,-pi,pi),(y,-pi,pi))
# I don't know multi-dimension of sympy
```

## Scope

```python
from sympy.abc import *
from sympy import *
fourier_series(E**(-t),(t,-pi,pi)).truncate(4).simplify()
```

```python
from sympy.abc import *
from sympy import *
fourier_series(E**(-t**2),(t,-pi,pi)).truncate(3).simplify()
```

```python
from sympy.abc import *
from sympy import *
fourier_series(Abs(t),(t,-pi,pi)).truncate(2).rewrite(exp).simplify()
```

```python
from sympy.abc import *
from sympy import *
plot(Abs(t),fourier_series(Abs(t),(t,-pi,pi)).truncate(2),(t,-pi,pi))
```

```python
from sympy.abc import *
from sympy import *
fourier_series(E**(t*I),(t,-pi,pi)).truncate(3)
```

### Options

```python
from sympy.abc import *
from sympy import *
fourier_series(E**(-Abs(t)) ,(t,-Rational(1,2),Rational(1,2))).truncate(3).rewrite(exp).expand()
```
