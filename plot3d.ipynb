## Basic Examples

Plot a function:

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x = np.outer(np.linspace(-3, 3, 30), np.ones(30))
y = np.linspace(-2, 2, 30)
z = np.sin(x+  y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y,z)
plt.show()
```

Plot several functions:

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x = np.outer(np.linspace(-3, 3, 30), np.ones(30))
y = np.linspace(-2, 2, 30)
z = np.sin(x+  y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y,x**2+y**2)
ax.plot_surface(x,y,-x**2-y**2)
plt.show()
```

Label each curve:

```python
#  x^2+y^2<=4 
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x = np.outer(np.linspace(-3, 3, 30), np.ones(30))
y = np.linspace(-2, 2, 30)

z = np.sin(x+  y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y,x**2+y**2+0*np.sqrt(4-(x**2+y**2)))
ax.plot_surface(x,y,-x**2-y**2+0*np.sqrt(4-(x**2+y**2)))
plt.show()
```

```python
# 

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x = np.outer(np.linspace(-2, 2, 30), np.ones(30))
y = np.linspace(-2, 2, 30)
z = np.sin(x+  y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(
    x, y,np.imag(
        np.arcsin((x+1j*y)**4)
                 )
                )
plt.show()
```

Fill between two curves:

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x = np.outer(np.linspace(0, 3, 30), np.ones(30))
y = np.linspace(0, 3, 30)
z = np.sin(x*y)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(
    x, y,z 
                )
plt.show()
```
