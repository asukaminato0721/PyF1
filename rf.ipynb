### Integral Transforms

### Function Identities and Simplifications

```python
from sympy import *
a,n,m=symbols("a n m")
Eq(rf(a,n),rf(a,m)*rf(a+m,n-m)).simplify()
```

```python
from sympy import *
a,n,m=symbols("a n m")
Eq(rf(a,n),(a+n-1)/(a-1)*rf(a-1,n)).simplify()
```

```python
from sympy import *
a,n,m=symbols("a n m")
Eq(rf(a,n),a/(a+n)*rf(a+1,n)).simplify()
```
