# Examples

## Basic Examples

Plot a function:

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,6*np.pi,300)
plt.plot(x,np.sin(x))
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0, 6*np.pi, -1, 1])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position(('data', 0))
```

Plot several functions with a legend:

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,2*np.pi,100)
for i in [1,2,3]:
    plt.plot(x,np.sin(i*x))
    #plotlist.append(y)
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0, 2*np.pi, -1, 1])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.legend(["sin(x)","sin(2x)","sin(3x)"],loc='best')
ax.spines['bottom'].set_position(('data', 0))
```

Fill below a curve:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,15,100)
y = 2*np.sin(x)+x
plt.plot(x,y)
plt.fill_between(x,0,y, alpha = 0.2)
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0, 15, 0, 17.5])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

Fill between two curves:

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
y1=np.sin(x)+x/2
y2=np.sin(x)+x
plt.plot(x,y1)
plt.plot(x,y2)
plt.fill_between(x,y2,y1,alpha=0.2)
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0,10,0, 10])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```

Plot multiple filled curves, automatically using transparent colors:

```python
import scipy as sp
import scipy.special
import matplotlib.pyplot as plt
x=sp.linspace(0,10,100)
for i in range(1,5):
    y=sp.special.jv(i,x)
    plt.plot(x,y)
    plt.fill_between(x,0,y,alpha=0.15)
#---
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0, 10, -0.4, 0.6])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position(('data', 0))
#---
```

## Scope

### Sampling

More points are sampled when the function changes quickly:

```python
import scipy as sp
import matplotlib.pyplot as plt
x=sp.linspace(0,10,500)
plt.plot(x,sp.sin(30*sp.sin(x)))
#---
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0, 10, -1, 1])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position(('data', 0))
#---
```

The plot range is selected automatically:

```python
import scipy as sp
import matplotlib.pyplot as plt
x=sp.linspace(-1,1,500)
y=1/x
y[:-1][sp.diff(y)>0]=sp.nan
plt.plot(x,y)
plt.ylim((-10,10))
#---
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([-1, 1, -10, 10])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
#---
```

Ranges where the function becomes nonreal are excluded:

```python
import scipy as sp
import matplotlib.pyplot as plt
x=sp.linspace(0,1,500)
plt.plot(x,sp.sqrt(x))
#---
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0, 1, 0, 1])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
#---
```

The curve is split when there are discontinuities in the function:

```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
x=sp.linspace(0,100,500)
y=sp.floor(sp.sqrt(x))
y[:-1][np.abs(sp.diff(y))>0.3]=sp.nan
plt.plot(x,y)
ax=plt.gca()
xmin, xmax, ymin, ymax = plt.axis([0, 100, 0, 10])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
