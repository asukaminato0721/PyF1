#### 没有动画功能

```python
from sympy import *
t,ω=symbols('t ω')
x=symbols('x',cls=Function)
dsolve(
    Eq(
        Derivative(x(t),t,t)+ω**2*x(t),
        0
        )
    ).rewrite(cos)
```

#### 没有旋转矩阵功能

```python
from sympy import *
from sympy.plotting import *
Φ,θ=symbols("Φ θ")
plot3d_parametric_surface(
    cos(Φ)*sin(θ), 
    sin(θ)*sin(Φ),
    cos(θ),
    (Φ, -pi,pi), (θ, -pi, pi)
)
```

```python
from sympy import *
from sympy.plotting import *
Φ,θ=symbols("Φ θ")
plot3d_parametric_surface(
    cos(Φ)+1/2*cos(θ)*cos(Φ),
    sin(Φ)+1/2*cos(θ)*sin(Φ),
    sin(θ)/2,
    (Φ, -pi, pi), (θ, 0, 2*pi)
    )
```
