## Basic Examples

```python
y=Function("y")
dsolve(Eq(y(x).diff(x) + y(x) , a *sin(x)))
```

```python
y=Function("y")
dsolve(Eq(y(x).diff(x) + y(x) , a *sin(x)),
      ics={
          y(0):0
      }      
      )
```

```python
y=Function("y")
dsolve(Eq(y(x).diff(x) ,3* y(x))    
      )
```

```python
y=Function("y")
dsolve(Eq(y(x).diff(x) ,3* y(x)) ,
       ics={
           y(0):5
       }
      )
```

```python
y=Function("y")
dsolve(Eq(y(x).diff(x) ,- y(x)) 
      
      )
```

```python
y=Function("y")
dsolve(Eq(y(x).diff(x) ,- 3*y(x)**2) ,
      ics={
          y(1):2
      }
      )
```

```python
y=Function("y")
dsolve(
    Eq(
        y(x).diff(x) , y(x)+a*cos(x)
    ) ,
      ics={
          y(0):3
      }
      )
```

```python
y=Function("y")
dsolve(
    Eq(
        y(x).diff(x,x) +4*y(x), 7
    ) 
      )
```

```python
y=Function("y")
dsolve(
    Eq(
        y(x).diff(x,x) +4*y(x), 7
    ) ,ics={
        y(x).diff(x).subs(x, 0): 2,
        y(0):1
    }
      )
```

```python
y=Function("y")
z=Function("z")
dsolve(
    [Eq(
        y(x).diff(x) , y(x)-a*z(x)
    ),
    Eq(
        z(x).diff(x),y(x)-z(x)
    )],ics={
        z(0):4,
        y(0):1
    }
      )
```

```python
sol = dsolve(
    [
        Eq(y(x).diff(x) - 3* z(x),sin (x   )),
        Eq(y(x) + z(x) , Rational(1,5))
    ], 
   ics={
       y(pi/2) : Rational(1,2)
   }
)
```

```python
from sympy.solvers.pde import pdsolve
u=Function("u")
y,x=symbols("y x")
eqn = diff(u(x, y), x) + 3 *diff(u(x, y), y) + u(x, y) - 1
pdsolve(eqn)
```
